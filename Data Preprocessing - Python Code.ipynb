```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# 1. Load the logically synced dataset
df = pd.read_csv(r"C:\Users\Nidish Kumaar V\OneDrive\Viknesh\GUVI - Data Science Course Materials\Capstone Projects\NutriClass Food Classification Using Nutritional Data Project\NutriClass_Logic_Fixed.csv")

# 2. Remove duplicates
df = df.drop_duplicates().reset_index(drop=True)

# 3. Impute missing values with Median
numerical_cols = ['Calories', 'Protein', 'Fat', 'Carbs', 'Sugar', 'Fiber', 
                  'Sodium', 'Cholesterol', 'Glycemic_Index', 'Water_Content', 'Serving_Size']
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())

# 4. Cap Outliers using IQR
def cap_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    return series.clip(lower=Q1 - 1.5 * IQR, upper=Q3 + 1.5 * IQR)

for col in numerical_cols:
    df[col] = cap_outliers(df[col])

# 5. Standardize Features
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Save the final preprocessed version
df.to_csv('NutriClass_Step1_Complete.csv', index=False)
```
